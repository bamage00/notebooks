```python
import simfin as sf
import pandas as pd
import matplotlib.pyplot as plt

from simfin.names import *
```

```python
sf.set_api_key('free')
sf.set_data_dir('D://Notebooks/data/simfin_data/')
```

```python
num_quarters = 4*10
num_days = 365*10
fig_size_1 = (8, 8)
```

```python
df_income = sf.load_income(variant='quarterly', market='us').reset_index()
df_income.columns
```

```python
df_cash = sf.load_cashflow(variant='quarterly', market='us').reset_index()
df_cash.columns
```

```python
df_balance = sf.load_balance(variant='quarterly', market='us').reset_index()
df_balance.columns
```

```python
df_prices = sf.load_shareprices(variant='daily', market='us').reset_index()
df_prices['Market Cap'] = df_prices['Close'] * df_prices['Shares Outstanding']
df_prices.columns
```

```python
tickers = ['AAPL', 'AMZN', 'GOOG', 'MSFT', 'FB', 'UBER', 'XOM']

metrics_income = ['Revenue', 'Net Income']

for m in metrics_income:    
    fig, axes = plt.subplots(sharex=True)
    for t in tickers:
        df_income[df_income['Ticker'] == t].tail(num_quarters).plot(
            x='Report Date', y=m, style='-', grid=True, figsize=fig_size_1, ax=axes, title=m)
    axes.legend(tickers)
    
    
metrics_balance = ['Cash, Cash Equivalents & Short Term Investments', 'Short Term Debt', 'Long Term Debt',
           'Retained Earnings', 'Total Equity']

for m in metrics_balance:    
    fig, axes = plt.subplots(sharex=True)
    for t in tickers:
        df_balance[df_balance['Ticker'] == t].tail(num_quarters).plot(
            x='Report Date', y=m, style='-', grid=True, figsize=fig_size_1, ax=axes, title=m)
    axes.legend(tickers)
    
metrics_prices = ['Dividend', 'Shares Outstanding', 'Market Cap']

for m in metrics_prices:    
    fig, axes = plt.subplots(sharex=True)
    for t in tickers:
        s = '.' if m == 'Dividend' else '-'
        df_prices[df_prices['Ticker'] == t].tail(num_days).plot(
            x='Date', y=m, grid=True, figsize=fig_size_1, style=s, ax=axes, title=m)
    axes.legend(tickers)
```
