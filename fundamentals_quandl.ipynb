```python
metrics = ['marketcap', 'revenue', 'netinc', 'eps', 'capex', 'cashneq', 'equity', 'debt', 'debtnc', 'ebitda',
           'roe', 'roa', 'fcf', 'gp', 'opinc', 'netmargin', 'pe', 'ps', 'pb', 'de', 'divyield']

for m in metrics:
    fig, axes = plt.subplots(sharex=True)
    data[data['ticker'].isin(tickers)].groupby('ticker').plot(
        x='calendardate', y=[m], style=['-'], kind='line', grid=True, figsize=fig_size_1, title=m, ax=axes)
    axes.legend(sorted(tickers))
```
